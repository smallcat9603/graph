```python
import igraph as ig
import json
```

```python
g = ig.Graph(directed=False)
with open("data/data.json", "r") as file:
    data = json.load(file)
for key in data:
    g.add_vertex(name=key)
    for element0 in data[key]:
        for element1 in data[key][element0]:
            node = element0 + "-" + element1
            if "tfidf" not in data[key][element0][element1]:
                # print(key, " ", data[key], " ", element0, " ", element1)
                g.add_vertex(name=node, lines=data[key][element0][element1]["lines"])
            else:
                g.add_vertex(name=node, lines=data[key][element0][element1]["lines"], tfidf=data[key][element0][element1]["tfidf"])
            g.add_edge(key, node)
```

```python
ig.summary(g)
```

```python
visual_style = {
    "vertex_label": g.vs["name"]
}
ig.plot(g, **visual_style)
```
